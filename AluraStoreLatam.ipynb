#1. Análisis de Facturación
* Facturación total de cada tienda, suma de ingreso total, y del ingreso anual por tienda. Cuál tienda factura menos ?)


```python
#print(f'\nFacturación total de cada tienda en Lista: ')

# DataFrames: tiendas
Precio = tienda['Precio']

facturacion_tienda = [float(tienda['Precio'].sum()),float(tienda2['Precio'].sum()),float(tienda3['Precio'].sum()),float(tienda4['Precio'].sum())]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Encontrar el valor mínimo de facturación
facturacion_minima = min(facturacion_tienda)


# Encontrar el índice del valor mínimo
facturacion_minima_indice = facturacion_tienda.index(facturacion_minima)

# Imprimir el resultado
print(f'\n-------------------------------------------------------------------------------------')
print(f'\n**La tienda que tiene menor facturación es la tienda # {facturacion_minima_indice + 1} con un valor de ${facturacion_minima:,.2f}**')

# Visualización facturación de cada tienda
print(f'\n-------------------------------------------------------------------------------------')
print(f'\nFacturación total de cada tienda: ')
print(f'\n')

for i in range(0,len(facturacion_tienda)):
  print(f"\t- Tienda {i+1} es: ${facturacion_tienda[i]:,.2f}")
print(f'\n-------------------------------------------------------------------------------------')
```

* Facturacion por tienda a lo largo de los años

```python
import matplotlib.pyplot as plt
import pandas as pd

# DataFrames: tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Concatenar todos los DataFrames en uno solo
all_data = pd.concat(tiendas, keys=nombres_tiendas)
all_data = all_data.reset_index(level=0)
all_data = all_data.rename(columns={'level_0': 'Tienda'}) # Renombrar la columna 'level_0' a 'Tienda'

# Extraer el año de la columna 'Fecha de Compra'
all_data['Año'] = pd.to_datetime(all_data['Fecha de Compra']).dt.year

# Agrupar por tienda y año, sumando las ventas
ventas_por_tienda_año = all_data.groupby(['Tienda', 'Año'])['Precio'].sum().reset_index()

# Crear el gráfico de líneas
fig, ax = plt.subplots(figsize=(10, 6))

for tienda in nombres_tiendas:
    data_tienda = ventas_por_tienda_año[ventas_por_tienda_año['Tienda'] == tienda]
    ax.plot(data_tienda['Año'], data_tienda['Precio'], label=tienda)

# Personalizar el gráfico
ax.set_xlabel('Año')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por tienda a lo largo de los años')
ax.legend()
ax.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Configurar el formato de visualización de los números float
pd.options.display.float_format = '{:.2f}'.format

# Crear tabla resumen con valores por año y total
tabla_resumen = ventas_por_tienda_año.pivot(index='Tienda', columns='Año', values='Precio').fillna(0)  # Llenar valores NaN con 0
tabla_resumen['Total'] = tabla_resumen.sum(axis=1)  # Agregar columna Total
tabla_resumen = tabla_resumen.astype(float)  # Convertir a tipo float
print("\nTabla Resumen:")
print(tabla_resumen)
```

**Analisis** : Aunque la Tienda 3 sigue teniendo un rendimiento menor en varios aspectos, la Tienda 4 es la que ha experimentado la mayor caída en las ventas en los últimos años, como se puede apreciar en la gráfica.

Por lo tanto, sería importante revisar con más detalle la situación de la Tienda 4 y considerar si esta caída en las ventas es un factor determinante en la decisión de cierre,

# 4. Productos más y menos vendidos
## (Cuáles son los productos más y menos vendidos en cada tienda)

```python
tiendas = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas

for i, tienda_df in enumerate(tiendas):
    # Agrupar por producto y contar las ventas
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')

    # Ordenar por cantidad vendida (descendente para el más vendido, ascendente para el menos vendido)
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]

    # Imprimir los resultados
    print(f"Tienda {i + 1}:")
    print(f"  Producto más vendido: {mas_vendido['Producto']} ({mas_vendido['Cantidad Vendida']} ventas)")
    print(f"  Producto menos vendido: {menos_vendido['Producto']} ({menos_vendido['Cantidad Vendida']} ventas)")
    print("-" * 20)  # Separador entre tiendas
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener todas las categorías únicas
all_categories = pd.unique(pd.concat([tienda_df['Categoría del Producto'] for tienda_df in tiendas]))

# Crear un diccionario para almacenar los datos de cada categoría por tienda
category_data = {}
for category in all_categories:
    category_data[category] = [tienda_df['Categoría del Producto'].value_counts().get(category, 0) for tienda_df in tiendas]

# Crear el gráfico de barras apiladas
fig, ax = plt.subplots()
bottom = np.zeros(len(nombres_tiendas))

for category, data in category_data.items():
    ax.bar(nombres_tiendas, data, bottom=bottom, label=category)
    bottom += data

# Personalizar el gráfico
ax.set_ylabel('Cantidad de productos')
ax.set_title('Distribución de categorías de productos por tienda')
ax.legend(title='Categorías', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x si es necesario
plt.tight_layout()
plt.show()
```

la distribución de categorías de productos en cada tienda de forma unificada. 4. Productos mas y menos vendidos

```python
import matplotlib.pyplot as plt
import numpy as np

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

productos_mas_vendidos = []
productos_menos_vendidos = []

for tienda_df in tiendas:
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]['Producto']
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]['Producto']
    productos_mas_vendidos.append(mas_vendido)
    productos_menos_vendidos.append(menos_vendido)

# Crear el gráfico de barras agrupadas
x = np.arange(len(nombres_tiendas))  # Posiciones de las barras en el eje x
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, productos_mas_vendidos, width, label='Más Vendido', color='skyblue')
rects2 = ax.bar(x + width/2, productos_menos_vendidos, width, label='Menos Vendido', color='coral')

# Personalizar el gráfico
ax.set_ylabel('Productos')
ax.set_title('Productos Más y Menos Vendidos por Tienda')
ax.set_xticks(x)
ax.set_xticklabels(nombres_tiendas)
ax.legend()

# Agregar etiquetas de datos a las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```

 la tienda que tiene los productos menos vendidos con menor cantidad de ventas en comparación con las demás

Un costo de envío promedio alto puede ser un factor que influya en la decisión de los clientes al momento de comprar.

Si una tienda tiene un costo de envío promedio significativamente más alto que las demás, podría ser una señal de que la tienda está ubicada en una zona con mayor dificultad para la entrega o que la tienda tiene mayores costos de logística. Esta información puede ser relevante para tomar una decisión sobre qué tienda cerrar.

Si la tienda con el costo de envío promedio más alto también tiene un bajo rendimiento en ventas y calificación de los clientes, entonces se convierte en una fuerte candidata para el cierre.
