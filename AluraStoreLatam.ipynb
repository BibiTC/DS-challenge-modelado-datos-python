#1. Análisis de Facturación
* Facturación total de cada tienda, suma de ingreso total, y del ingreso anual por tienda. Cuál tienda factura menos ?)


```python
#print(f'\nFacturación total de cada tienda en Lista: ')

# DataFrames: tiendas

facturacion_tienda = [float(tienda['Precio'].sum()),float(tienda2['Precio'].sum()),float(tienda3['Precio'].sum()),float(tienda4['Precio'].sum())]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Encontrar el valor mínimo de facturación
facturacion_minima = min(facturacion_tienda)


# Encontrar el índice del valor mínimo
facturacion_minima_indice = facturacion_tienda.index(facturacion_minima)

# Imprimir el resultado
print(f'\n-------------------------------------------------------------------------------------')
print(f'\n**La tienda que tiene menor facturación es la tienda # {facturacion_minima_indice + 1} con un valor de ${facturacion_minima:,.2f}**')

# Visualización facturación de cada tienda
print(f'\n-------------------------------------------------------------------------------------')
print(f'\nFacturación total de cada tienda: ')
print(f'\n')

for i in range(0,len(facturacion_tienda)):
  print(f"\t- Tienda {i+1} es: ${facturacion_tienda[i]:,.2f}")
print(f'\n-------------------------------------------------------------------------------------')
```

* Facturacion por tienda a lo largo de los años

```python
import matplotlib.pyplot as plt
import pandas as pd

# DataFrames: tiendas
tiendas_bd = [tienda, tienda2, tienda3, tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Concatenar todos los DataFrames en uno solo
all_data = pd.concat(tiendas_bd, keys=tiendas)
all_data = all_data.reset_index(level=0)
all_data = all_data.rename(columns={'level_0': 'Tienda'}) # Renombrar la columna 'level_0' a 'Tienda'

# Extraer el año de la columna 'Fecha de Compra'
all_data['Año'] = pd.to_datetime(all_data['Fecha de Compra']).dt.year

# Agrupar por tienda y año, sumando las ventas
ventas_por_tienda_año = all_data.groupby(['Tienda', 'Año'])['Precio'].sum().reset_index()

# Crear el gráfico de líneas
fig, ax = plt.subplots(figsize=(10, 6))

for tienda in tiendas:
    data_tienda = ventas_por_tienda_año[ventas_por_tienda_año['Tienda'] == tienda]
    ax.plot(data_tienda['Año'], data_tienda['Precio'], label=tienda)

# Personalizar el gráfico
ax.set_xlabel('Año')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por tienda a lo largo de los años')
ax.legend()
ax.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Configurar el formato de visualización de los números float
pd.options.display.float_format = '{:.2f}'.format

# Crear tabla resumen con valores por año y total
tabla_resumen = ventas_por_tienda_año.pivot(index='Tienda', columns='Año', values='Precio').fillna(0)  # Llenar valores NaN con 0
tabla_resumen['Total'] = tabla_resumen.sum(axis=1)  # Agregar columna Total
tabla_resumen = tabla_resumen.astype(float)  # Convertir a tipo float
print("\nTabla Resumen:")
print(tabla_resumen)
```

**Analisis** : Aunque la Tienda 3 sigue teniendo un rendimiento menor en varios aspectos, la Tienda 4 es la que ha experimentado la mayor caída en las ventas en los últimos años, como se puede apreciar en la gráfica.

# 2. Ventas por categoría
* Evidenciar las categorías más vendidas en cada tienda

* Analizar cantidad de ventas por cada tienda agrupadas por categoría.

```python
#tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
#tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener la lista de categorías no repetidas
categorias = set()
for tienda_df in tiendas_bd:  # Usar tiendas_bd en lugar de tiendas
    categorias.update(tienda_df['Categoría del Producto'].unique())
categorias = sorted(list(categorias))  # Ordenar alfabéticamente

# Crear un diccionario para almacenar los datos
datos_por_categoria = {}
for categoria in categorias:
    datos_por_categoria[categoria] = {}
    for i, tienda_df in enumerate(tiendas_bd):  # Usar tiendas_bd en lugar de tiendas
        # Filtrar por categoría y calcular ventas y cantidad de vendidas
        ventas_categoria = tienda_df[tienda_df['Categoría del Producto'] == categoria]['Precio'].sum()
        cantidad_vendida = tienda_df[tienda_df['Categoría del Producto'] == categoria]['Producto'].count()
        datos_por_categoria[categoria][tiendas[i]] = {'Ventas': ventas_categoria, 'Cantidad Vendida': cantidad_vendida}  # Usar tiendas para los nombres de las tiendas
```

```python
from prettytable import PrettyTable

#tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
#tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear la tabla con PrettyTable
tabla = PrettyTable()
tabla.field_names = ["Categoría"] + [f"{tienda} {tipo}" for tienda in tiendas for tipo in  ["Ventas", "Cantidad Vendida"]]

# Agregar filas a la tabla con estilo de líneas alternadas
for i, (categoria, datos_tiendas) in enumerate(datos_por_categoria.items()):
    fila = [categoria]
    for tienda in tiendas:
        fila.extend([datos_tiendas[tienda]['Ventas'], datos_tiendas[tienda]['Cantidad Vendida']])
    tabla.add_row(fila)


tabla.add_row(["", "", "", "", "", "", "", "", ""])  # Agregar una fila vacía para el espacio gris (opcional)

# Agregar fila TOTAL
fila_total = ["TOTAL"]
for tienda in tiendas:
    total_ventas = sum(datos_tiendas[tienda]['Ventas'] for datos_tiendas in datos_por_categoria.values())
    total_cantidad_vendida = sum(datos_tiendas[tienda]['Cantidad Vendida'] for datos_tiendas in datos_por_categoria.values())
    fila_total.extend([total_ventas, total_cantidad_vendida])
tabla.add_row(fila_total)

# Imprimir la tabla
print(tabla)
```

# 3. Calificación promedio de la tienda
## Analizar el promedio de evaluación de los clientes para cada tienda

```python
tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, tienda_df in enumerate(tiendas_bd):
    # Asegúrate de que los datos de la columna "Calificación" sean de tipo numérico
    tienda_df["Calificación"] = pd.to_numeric(tienda_df["Calificación"], errors='coerce')

    # Calcular la calificación promedio manualmente
    suma_calificaciones = 0
    num_calificaciones = 0
    for calificacion in tienda_df["Calificación"]:
        if not pd.isna(calificacion):  # Ignorar valores NaN
            suma_calificaciones += calificacion
            num_calificaciones += 1

    calificacion_promedio = suma_calificaciones / num_calificaciones if num_calificaciones else 0  # Evitar división por cero

    # Imprimir el resultado
    print(f"Tienda {i + 1}: Calificación promedio = {calificacion_promedio:.2f}")
```

```python
import matplotlib.pyplot as plt

tiendas_bd = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

calificaciones_promedio = []
for tienda_df in tiendas_bd:
    calificaciones_promedio.append(tienda_df["Calificación"].mean())

# Crear el gráfico de barras
plt.bar(tiendas, calificaciones_promedio, color=['blue', 'green', 'red', 'orange'])

# Personalizar el gráfico
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(3.0, 4.5)  # Ajustar el rango del eje y para mejor visualización
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

- La tienda 3 tiene la calificación promedio mas alta.

```python
import matplotlib.pyplot as plt
import pandas as pd

# DataFrames: tiendas
tiendas_bd = [tienda, tienda2, tienda3, tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Concatenar todos los DataFrames en uno solo
all_data = pd.concat(tiendas_bd, keys=tiendas)
all_data = all_data.reset_index(level=0)
all_data = all_data.rename(columns={'level_0': 'Tienda'}) # Renombrar la columna 'level_0' a 'Tienda'

# Extraer el año de la columna 'Fecha de Compra'
all_data['Año'] = pd.to_datetime(all_data['Fecha de Compra']).dt.year

# Agrupar por tienda y año, calculando la calificación promedio
calificaciones_promedio = all_data.groupby(['Tienda', 'Año'])['Calificación'].mean().reset_index()

# Crear el gráfico de líneas
fig, ax = plt.subplots(figsize=(10, 6))

for tienda in tiendas:
    data_tienda = calificaciones_promedio[calificaciones_promedio['Tienda'] == tienda]
    ax.plot(data_tienda['Año'], data_tienda['Calificación'], label=tienda)

# Personalizar el gráfico
ax.set_xlabel('Año')
ax.set_ylabel('Calificación Promedio')
ax.set_title('Calificación Promedio por Tienda a lo largo de los años')
ax.legend()
ax.grid(True)

# Mostrar el gráfico
plt.show()
```

# 4. Productos más y menos vendidos en cada tienda

```python
# DataFrames: tiendas
tiendas_bd = [tienda, tienda2, tienda3, tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for i, tienda_df in enumerate(tiendas_bd):
    # Agrupar por producto y contar las ventas
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')

    # Ordenar por cantidad vendida (descendente para el más vendido, ascendente para el menos vendido)
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]

    # Obtener la categoría del producto más vendido y menos vendido
    categoria_mas_vendido = tienda_df[tienda_df['Producto'] == mas_vendido['Producto']]['Categoría del Producto'].iloc[0]
    categoria_menos_vendido = tienda_df[tienda_df['Producto'] == menos_vendido['Producto']]['Categoría del Producto'].iloc[0]

    # Calcular la calificación promedio manualmente
    suma_calificaciones = 0
    num_calificaciones = 0
    for calificacion in tienda_df["Calificación"]:
        if not pd.isna(calificacion):  # Ignorar valores NaN
            suma_calificaciones += calificacion
            num_calificaciones += 1

    calificacion_promedio = suma_calificaciones / num_calificaciones if num_calificaciones else 0  # Evitar división por cero

    # Imprimir los resultados
    print(f"Tienda {i + 1}:")
    print(f"  Producto más vendido: {mas_vendido['Producto']} ({mas_vendido['Cantidad Vendida']} ventas) - Categoría: {categoria_mas_vendido}")
    print(f"  Producto menos vendido: {menos_vendido['Producto']} ({menos_vendido['Cantidad Vendida']} ventas) - Categoría: {categoria_menos_vendido}")
    print(f"  Calificación promedio: {calificacion_promedio:.2f}")  # Imprimir la calificación promedio
    print("-" * 20)  # Separador entre tiendas
```

 - Al analizar esta información nos damos cuenta que en la Tienda 2 los productos mas vendidos estan en la categoria de Libros y Juguetes, que son categorias que no estan generando gran facturacion en valores.
 - Si solo validamos este item podriamos pensar en que la tienda 2 es la tienda menos rentable, y opcionada a ser cerrada.

 ## Grafico de Burbujas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# DataFrames: tiendas
tiendas_bd = [tienda, tienda2, tienda3, tienda4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener todas las categorías únicas
categorias = set()
for tienda_df in tiendas_bd:  # Usar tiendas_bd en lugar de tiendas
    categorias.update(tienda_df['Categoría del Producto'].unique())
categorias = sorted(list(categorias))  # Ordenar alfabéticamente

# Crear un diccionario para almacenar los datos
datos_por_categoria = {}
for categoria in categorias:
    datos_por_categoria[categoria] = {}
    for i, tienda_df in enumerate(tiendas_bd):  # Usar tiendas_bd en lugar de tiendas
        # Filtrar por categoría y calcular ventas y cantidad de vendidas
        ventas_categoria = tienda_df[tienda_df['Categoría del Producto'] == categoria]['Precio'].sum()
        cantidad_vendida = tienda_df[tienda_df['Categoría del Producto'] == categoria]['Producto'].count()
        datos_por_categoria[categoria][tiendas[i]] = {'Ventas': ventas_categoria, 'Cantidad Vendida': cantidad_vendida}  # Usar tiendas para los nombres de las tiendas
# Obtener productos más y menos vendidos
productos_mas_vendidos = []
productos_menos_vendidos = []
for tienda_df in tiendas_bd:
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]['Producto']
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]['Producto']
    productos_mas_vendidos.append(mas_vendido)
    productos_menos_vendidos.append(menos_vendido)

# Obtener calificaciones promedio
calificaciones_promedio = []
for tienda_df in tiendas_bd:
    calificacion_promedio = tienda_df["Calificación"].mean()
    calificaciones_promedio.append(calificacion_promedio)


# Crear el gráfico de burbujas
fig, ax = plt.subplots(figsize=(10, 6))

# Productos más vendidos
ax.scatter(productos_mas_vendidos, calificaciones_promedio, s=[datos_por_categoria[categoria][tienda]['Cantidad Vendida'] for categoria in categorias for tienda in tiendas],
           alpha=0.5, label='Más Vendidos')

# Productos menos vendidos
ax.scatter(productos_menos_vendidos, calificaciones_promedio, s=[datos_por_categoria[categoria][tienda]['Cantidad Vendida'] for categoria in categorias for tienda in tiendas],
           alpha=0.5, label='Menos Vendidos', marker='x')

# Personalizar el gráfico
ax.set_xlabel('Productos')
ax.set_ylabel('Calificación Promedio')
ax.set_title('Productos Más y Menos Vendidos por Calificación Promedio y Cantidad Vendida')
ax.legend()
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x si es necesario
plt.tight_layout()
plt.show()
```

Gráfico 2: Distribución de categorías de productos (Gráfico de pastel)

```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Crear una cuadrícula de 2x2 para los gráficos

for i, tienda_df in enumerate(tiendas):
    # Calcular la distribución de categorías de productos para la tienda actual
    categorias = tienda_df['Categoría del Producto'].value_counts().index
    cantidades = tienda_df['Categoría del Producto'].value_counts().values

    # Obtener el subplot correspondiente
    ax = axes[i // 2, i % 2]  # Calcular la posición del subplot en la cuadrícula

    # Crear el gráfico de pastel para la tienda actual
    ax.pie(cantidades, labels=categorias, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Distribución de Categorías de Productos ({nombres_tiendas[i]})')
    ax.axis('equal')  # Para que el gráfico sea circular

plt.tight_layout()  # Ajustar el espacio entre los subplots
plt.show()  # Mostrar el gráfico
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener todas las categorías únicas
all_categories = pd.unique(pd.concat([tienda_df['Categoría del Producto'] for tienda_df in tiendas]))

# Crear un diccionario para almacenar los datos de cada categoría por tienda
category_data = {}
for category in all_categories:
    category_data[category] = [tienda_df['Categoría del Producto'].value_counts().get(category, 0) for tienda_df in tiendas]

# Crear el gráfico de barras apiladas
fig, ax = plt.subplots()
bottom = np.zeros(len(nombres_tiendas))

for category, data in category_data.items():
    ax.bar(nombres_tiendas, data, bottom=bottom, label=category)
    bottom += data

# Personalizar el gráfico
ax.set_ylabel('Cantidad de productos')
ax.set_title('Distribución de categorías de productos por tienda')
ax.legend(title='Categorías', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x si es necesario
plt.tight_layout()
plt.show()
```

la distribución de categorías de productos en cada tienda de forma unificada. Podrás comparar la proporción de cada categoría en cada tienda, así como la cantidad total de productos de cada categoría en todas las tiendas. Esto te ayudará a identificar las tiendas con una distribución de categorías de productos menos diversificada o con un enfoque en categorías menos populares, lo cual podría ser un factor a considerar al decidir qué tienda cerrar.

```python
import matplotlib.pyplot as plt
import numpy as np

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

productos_mas_vendidos = []
productos_menos_vendidos = []

for tienda_df in tiendas:
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]['Producto']
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]['Producto']
    productos_mas_vendidos.append(mas_vendido)
    productos_menos_vendidos.append(menos_vendido)

# Crear el gráfico de barras agrupadas
x = np.arange(len(nombres_tiendas))  # Posiciones de las barras en el eje x
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, productos_mas_vendidos, width, label='Más Vendido', color='skyblue')
rects2 = ax.bar(x + width/2, productos_menos_vendidos, width, label='Menos Vendido', color='coral')

# Personalizar el gráfico
ax.set_ylabel('Productos')
ax.set_title('Productos Más y Menos Vendidos por Tienda')
ax.set_xticks(x)
ax.set_xticklabels(nombres_tiendas)
ax.legend()

# Agregar etiquetas de datos a las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```
