#1. An√°lisis de Facturaci√≥n
* Facturaci√≥n total de cada tienda, suma de ingreso total, y del ingreso anual por tienda. Cu√°l tienda factura menos ?)


```python
#print(f'\nFacturaci√≥n total de cada tienda en Lista: ')

# DataFrames: tiendas
Precio = tienda['Precio']

facturacion_tienda = [float(tienda['Precio'].sum()),float(tienda2['Precio'].sum()),float(tienda3['Precio'].sum()),float(tienda4['Precio'].sum())]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Encontrar el valor m√≠nimo de facturaci√≥n
facturacion_minima = min(facturacion_tienda)


# Encontrar el √≠ndice del valor m√≠nimo
facturacion_minima_indice = facturacion_tienda.index(facturacion_minima)

# Imprimir el resultado
print(f'\n-------------------------------------------------------------------------------------')
print(f'\n**La tienda que tiene menor facturaci√≥n es la tienda # {facturacion_minima_indice + 1} con un valor de ${facturacion_minima:,.2f}**')

# Visualizaci√≥n facturaci√≥n de cada tienda
print(f'\n-------------------------------------------------------------------------------------')
print(f'\nFacturaci√≥n total de cada tienda: ')
print(f'\n')

for i in range(0,len(facturacion_tienda)):
  print(f"\t- Tienda {i+1} es: ${facturacion_tienda[i]:,.2f}")
print(f'\n-------------------------------------------------------------------------------------')
```

* Facturacion por tienda a lo largo de los a√±os

```python
import matplotlib.pyplot as plt
import pandas as pd

# DataFrames: tiendas
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Concatenar todos los DataFrames en uno solo
all_data = pd.concat(tiendas, keys=nombres_tiendas)
all_data = all_data.reset_index(level=0)
all_data = all_data.rename(columns={'level_0': 'Tienda'}) # Renombrar la columna 'level_0' a 'Tienda'

# Extraer el a√±o de la columna 'Fecha de Compra'
all_data['A√±o'] = pd.to_datetime(all_data['Fecha de Compra']).dt.year

# Agrupar por tienda y a√±o, sumando las ventas
ventas_por_tienda_a√±o = all_data.groupby(['Tienda', 'A√±o'])['Precio'].sum().reset_index()

# Crear el gr√°fico de l√≠neas
fig, ax = plt.subplots(figsize=(10, 6))

for tienda in nombres_tiendas:
    data_tienda = ventas_por_tienda_a√±o[ventas_por_tienda_a√±o['Tienda'] == tienda]
    ax.plot(data_tienda['A√±o'], data_tienda['Precio'], label=tienda)

# Personalizar el gr√°fico
ax.set_xlabel('A√±o')
ax.set_ylabel('Ventas')
ax.set_title('Ventas por tienda a lo largo de los a√±os')
ax.legend()
ax.grid(True)

# Mostrar el gr√°fico
plt.show()
```

```python
# Configurar el formato de visualizaci√≥n de los n√∫meros float
pd.options.display.float_format = '{:.2f}'.format

# Crear tabla resumen con valores por a√±o y total
tabla_resumen = ventas_por_tienda_a√±o.pivot(index='Tienda', columns='A√±o', values='Precio').fillna(0)  # Llenar valores NaN con 0
tabla_resumen['Total'] = tabla_resumen.sum(axis=1)  # Agregar columna Total
tabla_resumen = tabla_resumen.astype(float)  # Convertir a tipo float
print("\nTabla Resumen:")
print(tabla_resumen)
```

Por lo tanto, ser√≠a importante revisar con m√°s detalle la situaci√≥n de la Tienda 4 y considerar si esta ca√≠da en las ventas es un factor determinante en la decisi√≥n de cierre,

# 3. Calificaci√≥n promedio de la tienda
## (Cu√°l es el promedio de evaluaci√≥n de cada uno de los clientes;)

```python
tiendas = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas

for i, tienda_df in enumerate(tiendas):
    # Calcular la calificaci√≥n promedio
    calificacion_promedio = tienda_df["Calificaci√≥n"].mean()

    # Imprimir el resultado
    print(f"Tienda {i + 1}: Calificaci√≥n promedio = {calificacion_promedio:.2f}")
```

# 4. Productos m√°s y menos vendidos
## (Cu√°les son los productos m√°s y menos vendidos en cada tienda)

```python
tiendas = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas

for i, tienda_df in enumerate(tiendas):
    # Agrupar por producto y contar las ventas
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')

    # Ordenar por cantidad vendida (descendente para el m√°s vendido, ascendente para el menos vendido)
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]

    # Imprimir los resultados
    print(f"Tienda {i + 1}:")
    print(f"  Producto m√°s vendido: {mas_vendido['Producto']} ({mas_vendido['Cantidad Vendida']} ventas)")
    print(f"  Producto menos vendido: {menos_vendido['Producto']} ({menos_vendido['Cantidad Vendida']} ventas)")
    print("-" * 20)  # Separador entre tiendas
```

Gr√°fico 2: Distribuci√≥n de categor√≠as de productos (Gr√°fico de pastel)

```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]  # Lista de DataFrames de las tiendas
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Crear una cuadr√≠cula de 2x2 para los gr√°ficos

for i, tienda_df in enumerate(tiendas):
    # Calcular la distribuci√≥n de categor√≠as de productos para la tienda actual
    categorias = tienda_df['Categor√≠a del Producto'].value_counts().index
    cantidades = tienda_df['Categor√≠a del Producto'].value_counts().values

    # Obtener el subplot correspondiente
    ax = axes[i // 2, i % 2]  # Calcular la posici√≥n del subplot en la cuadr√≠cula

    # Crear el gr√°fico de pastel para la tienda actual
    ax.pie(cantidades, labels=categorias, autopct='%1.1f%%', startangle=90)
    ax.set_title(f'Distribuci√≥n de Categor√≠as de Productos ({nombres_tiendas[i]})')
    ax.axis('equal')  # Para que el gr√°fico sea circular

plt.tight_layout()  # Ajustar el espacio entre los subplots
plt.show()  # Mostrar el gr√°fico
```

Este gr√°fico muestra la proporci√≥n de productos vendidos en cada categor√≠a para una tienda espec√≠fica (en este caso, la Tienda 1). Puedes modificar el c√≥digo para generar gr√°ficos similares para las otras tiendas.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Obtener todas las categor√≠as √∫nicas
all_categories = pd.unique(pd.concat([tienda_df['Categor√≠a del Producto'] for tienda_df in tiendas]))

# Crear un diccionario para almacenar los datos de cada categor√≠a por tienda
category_data = {}
for category in all_categories:
    category_data[category] = [tienda_df['Categor√≠a del Producto'].value_counts().get(category, 0) for tienda_df in tiendas]

# Crear el gr√°fico de barras apiladas
fig, ax = plt.subplots()
bottom = np.zeros(len(nombres_tiendas))

for category, data in category_data.items():
    ax.bar(nombres_tiendas, data, bottom=bottom, label=category)
    bottom += data

# Personalizar el gr√°fico
ax.set_ylabel('Cantidad de productos')
ax.set_title('Distribuci√≥n de categor√≠as de productos por tienda')
ax.legend(title='Categor√≠as', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')  # Rotar etiquetas del eje x si es necesario
plt.tight_layout()
plt.show()
```

4. Productos mas y menos vendidos

```python
import matplotlib.pyplot as plt
import numpy as np

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

productos_mas_vendidos = []
productos_menos_vendidos = []

for tienda_df in tiendas:
    ventas_por_producto = tienda_df.groupby('Producto')['Producto'].count().reset_index(name='Cantidad Vendida')
    mas_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=False).iloc[0]['Producto']
    menos_vendido = ventas_por_producto.sort_values('Cantidad Vendida', ascending=True).iloc[0]['Producto']
    productos_mas_vendidos.append(mas_vendido)
    productos_menos_vendidos.append(menos_vendido)

# Crear el gr√°fico de barras agrupadas
x = np.arange(len(nombres_tiendas))  # Posiciones de las barras en el eje x
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, productos_mas_vendidos, width, label='M√°s Vendido', color='skyblue')
rects2 = ax.bar(x + width/2, productos_menos_vendidos, width, label='Menos Vendido', color='coral')

# Personalizar el gr√°fico
ax.set_ylabel('Productos')
ax.set_title('Productos M√°s y Menos Vendidos por Tienda')
ax.set_xticks(x)
ax.set_xticklabels(nombres_tiendas)
ax.legend()

# Agregar etiquetas de datos a las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
```

 la tienda que tiene los productos menos vendidos con menor cantidad de ventas en comparaci√≥n con las dem√°s

Un costo de env√≠o promedio alto puede ser un factor que influya en la decisi√≥n de los clientes al momento de comprar.

Si una tienda tiene un costo de env√≠o promedio significativamente m√°s alto que las dem√°s, podr√≠a ser una se√±al de que la tienda est√° ubicada en una zona con mayor dificultad para la entrega o que la tienda tiene mayores costos de log√≠stica. Esta informaci√≥n puede ser relevante para tomar una decisi√≥n sobre qu√© tienda cerrar.

Si la tienda con el costo de env√≠o promedio m√°s alto tambi√©n tiene un bajo rendimiento en ventas y calificaci√≥n de los clientes, entonces se convierte en una fuerte candidata para el cierre.

```python

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

costos_envio_promedio = []

for tienda_df in tiendas:
    costo_envio_promedio = tienda_df['Costo de env√≠o'].mean()
    costos_envio_promedio.append(costo_envio_promedio)

# Crear el gr√°fico de barras
x = np.arange(len(nombres_tiendas))  # Posiciones de las barras en el eje x
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots()
rects = ax.bar(x, costos_envio_promedio, width, label='Costo de Env√≠o Promedio', color='skyblue')

# Personalizar el gr√°fico
ax.set_ylabel('Costo de Env√≠o Promedio')
ax.set_title('Costo de Env√≠o Promedio por Tienda')
ax.set_xticks(x)
ax.set_xticklabels(nombres_tiendas)
ax.legend()

# Agregar etiquetas de datos a las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{:.2f}'.format(height),  # Formatear con dos decimales
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects)

fig.tight_layout()
plt.show()
```
